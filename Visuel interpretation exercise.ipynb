## Importing / Installing packages



```python
# Packages / libraries
import os 
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


```

## 1. Loading the Raw Data


```python
sns.set(style="whitegrid", palette="pastel", color_codes=True)
sns.set(rc={'figure.figsize':(12,8)})
# Load the example tips dataset
tips = sns.load_dataset("tips")
tips
```

## Excercise

```python
#Barplots (Connection between tip and meal)
sns.set(rc={'figure.figsize':(9,4)})
ax = sns.barplot(x="time", y="tip", data=tips, ci=False)
```

```python
#Lineplot(tip and day)
ax = sns.lineplot(x='day', y='tip', data = tips, ci=None)
```

```python
#Lineplot(Connection between tip and day by sex)
ax = sns.lineplot(x='day', y='tip', hue="sex", data = tips, ci=None, markers=True)
```

```python
#Lineplot(Connection between total bill and tip)
ax = sns.lineplot(x='total_bill', y='tip', data = tips, ci=None, markers=True)
```

```python
#Linnar regression between total bill and tip
ax = sns.lmplot(x="total_bill", y="tip", data=tips, height=9)
```

```python
#Adding new column tip per person to see difference
tips["tip_per_person"] = tips["tip"]/tips["size"]
tips
```

```python
#Barplot by the sex with tips per person
sns.set(rc={'figure.figsize':(9,4)})
ax = sns.barplot(x="sex", y="tip_per_person", data=tips, ci=False)
```

```python
#Barplot sex and tip per person by day
sns.set(rc={'figure.figsize':(9,4)})
ax = sns.barplot(x="sex", y="tip_per_person",hue="day", data=tips, ci=False)
```

```python
#see difference when we use column tip instead tip per person 
sns.set(rc={'figure.figsize':(9,4)})
ax = sns.barplot(x="sex", y="tip",hue="day", data=tips, ci=False)
```

```python
#Distplot - distribution of tip
x = tips['tip'].values

sns.distplot(x, color = '#2BE911');

# Calculating the mean
mean = tips['tip'].mean()

#ploting the mean
plt.axvline(mean, 0,0.85, color = 'red')
```

```python
pc = tips[['smoker','tip','total_bill', 'size']].corr(method ='pearson')

cols = ['smoker CC','tip CC','total_bill CC']

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 40}, vmin=0, vmax=1)
```

```python
#Boxplot showing distribution of tip
x = tips['tip'].values

ax = sns.boxplot(x)
```

```python
#Pairplot of all datafrime
t = sns.pairplot(tips, plot_kws={'color':'blue'})
```
